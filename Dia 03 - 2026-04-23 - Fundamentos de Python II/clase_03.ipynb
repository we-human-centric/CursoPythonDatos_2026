Hoy damos el salto: scripts lineales a código organizado. Control de flujo (si-esto-entonces-aquello), estructuras de datos que almacenan múltiples valores (listas, tuplas, diccionarios), y funciones que reutilizas. Esta sesión es clave para todo lo que viene.

```python
import random
import numpy as np
import pandas as pd

SEED = 42
random.seed(SEED)
np.random.seed(SEED)

print("Setup completo ✓")
```

Condicionales. Si algo es verdad, haz X. Si no, haz Y. Base de la lógica de programación.

```python
# Guiado: calificación de examen
puntuacion = 85

if puntuacion >= 90:
    calificacion = "A"
elif puntuacion >= 80:
    calificacion = "B"
elif puntuacion >= 70:
    calificacion = "C"
else:
    calificacion = "F"

print(f"Puntuación: {puntuacion}, Calificación: {calificacion}")

assert calificacion == "B"
print("✓ Condicional bien")
```

Estructura: `if`, `elif` (else-if), `else`. Importante: la indentación (espacios) define qué código va dentro de cada bloque. Python lo toma en serio.

```python
# Bucles: for itera sobre colecciones
numeros = [1, 2, 3, 4, 5]

suma = 0
for num in numeros:
    suma += num  # suma = suma + num
    print(f"Número: {num}, suma acumulada: {suma}")

print(f"\nSuma final: {suma}")

assert suma == 15
print("✓ For loop bien")
```

`for` itera sobre cada elemento. `+=` es abreviatura de `suma = suma + num`. Las listas (como `[1, 2, 3, 4, 5]`) se crean con corchetes.

```python
# Listas, tuplas, sets, diccionarios
lista = [1, 2, 3, "texto"]          # Mutable, indexada, ordenada
tupla = (1, 2, 3)                   # Inmutable, indexada, ordenada
conjunto = {1, 2, 3, 2}             # Sin duplicados
diccionario = {"nombre": "Juan", "edad": 25}  # Clave: valor

print(f"Lista: {lista}")
print(f"Tupla: {tupla}")
print(f"Set (sin duplicados): {conjunto}")
print(f"Diccionario: {diccionario}")

# Acceso
print(f"\nElemento 0 de lista: {lista[0]}")
print(f"Nombre del diccionario: {diccionario['nombre']}")

assert lista[0] == 1
assert diccionario['nombre'] == "Juan"
print("✓ Estructuras de datos ok")
```

Listas: mutables (puedes cambiarlas), con duplicados. Tuplas: inmutables, más eficientes. Sets: sin duplicados, operaciones de conjuntos. Diccionarios: busca rápida con claves.

Recuerda que tuviste 500 vendedores ayer? Una lista resuelve eso; una lista de diccionarios resuelve datos complejos.

```python
# Aplicado: suma de lista de ventas
ventas = [150.50, 200.00, 75.25, 300.10, 125.75]

total = 0
for venta in ventas:
    total += venta

promedio = total / len(ventas)  # len() cuenta elementos

print(f"Ventas: {ventas}")
print(f"Total: ${total:.2f}")
print(f"Promedio: ${promedio:.2f}")

assert abs(total - 851.60) < 0.01
assert abs(promedio - 170.32) < 0.01
print("✓ Suma y promedio correctos")
```

Combinamos lista + bucle: iterar sobre datos y acumular. Patrón que verás todo el tiempo.

```python
# Funciones: código reutilizable
def saludar(nombre):
    """Devuelve un saludo personalizado."""
    return f"Hola, {nombre}. Bienvenido."

def calcular_promedio(calificaciones):
    """Calcula el promedio de una lista de notas."""
    return sum(calificaciones) / len(calificaciones)

# Usar las funciones
msg = saludar("María")
print(msg)

notas = [85, 90, 78, 92]
prom = calcular_promedio(notas)
print(f"Promedio de notas: {prom:.2f}")

assert msg == "Hola, María. Bienvenido."
assert abs(prom - 86.25) < 0.01
print("✓ Funciones funcionan")
```

Las funciones encapsulan lógica. El string entre triple comilla es el docstring (documentación). `return` devuelve el resultado. Una vez definidas, las reutilizas.

```python
# Desafío: función que valida email
def es_email_valido(email):
    """Valida que email tenga exactamente un @ y al menos un . después."""
    cuenta_arroba = email.count('@')
    posicion_arroba = email.find('@')
    posicion_punto = email.find('.', posicion_arroba)
    
    return (cuenta_arroba == 1) and (posicion_punto > posicion_arroba) and (posicion_punto != -1)

# Probar
emails_test = [
    ("juan@example.com", True),
    ("maria.perez@empresa.org", True),
    ("invalido@.com", False),
    ("sin-arroba.com", False),
]

for email, esperado in emails_test:
    resultado = es_email_valido(email)
    estado = "✓" if resultado == esperado else "✗"
    print(f"{estado} {email}: {resultado}")
    assert resultado == esperado

print("\n✓ Todas las validaciones correctas")
```
