## 1. Control de flujo · decisiones

Hasta ahora nuestros programas se ejecutan de arriba hacia abajo sin ramificaciones. El **control de flujo** cambia eso: permite ejecutar código solo cuando se cumpla una condición, o repetir código varias veces.

Un programa "real" es un árbol de decisiones y bucles. La estructura básica de una decisión en Python es:

```python
if condicion:
    # se ejecuta si condicion es True
elif otra_condicion:
    # se ejecuta si la primera es False y esta es True
else:
    # se ejecuta si ninguna de las anteriores
```

**Indentación obligatoria**: en Python los bloques no se delimitan con llaves sino con **4 espacios** de indentación. Esto hace al código más legible pero un espacio mal puesto es un error de sintaxis.

## "if" statements (si esto entonces aquello)
  

````python
def evaluate_temp(temp):
    # Establece un mensaje inicial
    message = "Normal temperature."
    # Actualiza el mensaje en caso de que la temperatura sea mayor que 38
    if temp > 38:
        message = "Fever!"
    return message
````

## "if ... else" statements

En español viene a ser **si** se cumple esto entonces haz esto, **sino** entonces haz esto otro.

````python
def evaluate_temp_with_else(temp):
    if temp > 38:
        message = "Fever!"
    else:
        message = "Normal temperature."
    return message
````

````python
print(evaluate_temp_with_else(12))
````

## "if ... elif ... else" statements
Puedes pensar en "elif" como si dijera ... "ya que esa condición anterior (p. ej., `temp > 38`) era falsa, entonces vamos a comprobar si esta nueva condición (p. ej., `temp > 35`) podría ser cierta."

````python
def evaluate_temp_with_elif(temp):
    if temp > 38:
        message = "Fever!"
    elif temp > 35:
        message = "Normal temperature."
    else:
        message = "Low temperature."
    return message
````

````python
evaluate_temp_with_elif(12)
````

## 2. Bucles · repetición

Los bucles permiten ejecutar un bloque de código muchas veces. Python tiene dos:

- **`for`** — recorre los elementos de una colección (lista, string, diccionario, rango...). Es el bucle *por defecto* en Python.
- **`while`** — repite mientras una condición sea verdadera. Se usa cuando no sabemos cuántas iteraciones habrá.

Dos instrucciones controlan el flujo dentro de los bucles:
- **`break`** — sale del bucle inmediatamente.
- **`continue`** — salta a la siguiente iteración sin ejecutar lo que queda.

**`range(n)`** es un generador perezoso que produce los enteros `0, 1, ..., n-1`. `range(a, b, paso)` permite especificar inicio, fin (exclusivo) y paso.

> "Un bucle mal diseñado es la causa más común de un programa que nunca termina." — anónimo, pero cierto.

# Bucles en Python

Los bucles son una parte importante de la programación, ya que permiten repetir un bloque de código varias veces. En Python, hay dos tipos de bucles: bucles for y bucles while. Veamos más de cerca cómo funciona cada uno de estos bucles y proporcionemos algunos ejemplos y casos de uso.

## Bucles for

Un bucle for te permite iterar sobre una secuencia de valores, como una lista o un rango de números. Aquí hay un ejemplo:

````python
# Imprimir cada elemento en una lista usando un bucle for
fruits = ["manzana", "platano", "uva"]
for fruit in fruits:
    print(fruit)
````

Otro caso de uso común para un bucle for es repetir un bloque de código un cierto número de veces. Puedes hacer esto usando la función range para crear una secuencia de números para iterar. Aquí hay un ejemplo:

````python
# Imprimir los números del 1 al 5 usando un bucle for
for i in range(1, 6):
    print(i)
````

En este ejemplo, el programa utiliza la función range para crear una secuencia de números del 1 al 5, que luego se iteran usando un bucle for. El bucle asigna cada número a la variable i, que luego se imprime en la consola.

# Bucles while

Un bucle while te permite repetir un bloque de código mientras una cierta condición sea verdadera. Aquí hay un ejemplo:

````python
# Imprimir los números del 1 al 5 usando un bucle while
i = 1
while i <= 5:
    print(i)
    i += 1
````

En este ejemplo, el programa inicializa una variable i a 1 y utiliza un bucle while para repetir un bloque de código mientras i sea menor o igual a 5. El bucle imprime el valor de i en la consola y luego incrementa i en 1.

Otro caso de uso común para un bucle while es leer entradas del usuario hasta que se cumpla una cierta condición. Aquí hay un ejemplo:

````python
# Leer la entrada del usuario hasta que ingresen 'quit'
while True:
    user_input = input("Enter a word (or 'quit' to exit): ")

    if user_input == "quit":
        break

    print(f"You entered: {user_input}")
````

# Bucles anidados

Puedes usar bucles anidados en Python para iterar sobre múltiples secuencias a la vez. Aquí hay un ejemplo:

````python
# Imprimir todas las combinaciones posibles de dos tiradas de dados
for i in range(1, 7):
    for j in range(1, 7):
        print(f"({i}, {j})")
````

# Sentencias de control de bucles

Python proporciona varias sentencias de control de bucles que te permiten modificar el comportamiento de un bucle. La sentencia break te permite salir de un bucle antes de tiempo, mientras que la sentencia continue te permite omitir ciertas iteraciones de un bucle. Aquí hay un ejemplo:

````python
# Imprimir los números del 1 al 10, pero saltar el 5 y salir del bucle si llegamos al 8

for i in range(1, 11):
    if i == 5:
        continue

    print(i)

    if i == 8:
        break
````

### 3.2 Diccionarios

Los **diccionarios** son la estructura más útil para el análisis de datos después de las listas. Asocian una **clave** (casi siempre un string o número) con un **valor** (cualquier cosa).

Internamente son tablas hash: acceder a un valor por su clave es O(1), independientemente del tamaño. Esto los hace ideales cuando la pregunta es "dame el registro de X".

```python
poblacion = {"Quito": 2_011_000, "Guayaquil": 2_750_000}
poblacion["Cuenca"] = 636_000    # añadir
poblacion.get("Loja", 0)         # acceder con valor por defecto
```

Desde Python 3.7 los diccionarios preservan el orden de inserción (antes no era garantizado). Por eso pandas, JSON y muchos formatos de serialización los usan tanto.

1. Nos permiten identificar cada elemento (valor) por una clave. **Importante:** Las claves son únicas, no pueden ser las mismas
2. Encerrar el diccionario entre llaves ```{ }```
3. Las parejas de clave y valor se separan con comas; mientras que, la clave y el valor se separan con dos puntos ```{clave1: valor1, clave2: valor2}```
4. Asignar el diccionario a una variable que la representará, utilizando el operador ```=```
5. Se accede a los mismos a través de las claves (no a través de índices de posición)

````python
# claves: Nombre ciudades
# valores: Numero de registros de participantes a un evento

registro_ciudad = {'Lima': 692, 'Quito': 312, 'Montevideo': 95, 'Californina': 153, 'Medellín': 485}
registro_ciudad
````

## 4. Funciones · abstracción y reuso

Una **función** encapsula un pedazo de lógica bajo un nombre. Es la herramienta más importante para mantener el código manejable: cualquier bloque que hagas dos veces, extráelo a una función.

El principio DRY (*Don't Repeat Yourself*) resume la idea. Una función bien diseñada:

- Hace **una sola cosa** (principio de responsabilidad única).
- Tiene un **nombre claro** en verbo: `calcular_iva`, `limpiar_datos`, `entrenar_modelo`.
- Tiene una **docstring** que explica qué hace, qué recibe y qué devuelve.

### Argumentos

- **Posicionales**: se pasan en orden.
- **Con nombre (kwargs)**: `f(qty=6, price=2.5)`. Más legibles cuando hay muchos argumentos.
- **Por defecto**: `def f(tax=0.12): ...`. Permite argumentos opcionales.
- **Variables**: `*args` para número indeterminado de posicionales, `**kwargs` para nombrados.

Una función sin `return` devuelve `None`.

# Empaquetado de Diccionarios de Argumentos

El operador de doble asterisco (**) en una definición de función de Python indica el empaquetado de diccionarios de argumentos. Los argumentos correspondientes en la llamada a la función se empaquetan en un diccionario.

````python
def f(**kwargs):
    for key, value in kwargs.items():
        print(f'{key} -> {value}')

f(a=1, b=2, c=3)
# Salida:
# a -> 1
# b -> 2
# c -> 3
````

# Desempaquetado de Diccionarios de Argumentos

Cuando el operador de doble asterisco (**) precede a un argumento en una llamada a función, esto indica que el argumento es un diccionario que debe desempaquetarse y pasarse a la función como argumentos con palabras clave.

````python
def f(a, b, c):
    print(a, b, c)

d = {'a': 1, 'b': 2, 'c': 3}
f(**d)
# Salida:
# 1 2 3
````

# Poniéndolo Todo Junto

Es posible usar parámetros posicionales estándar, *args, y **kwargs en una misma definición de función de Python. Deben especificarse en ese orden.

````python
def f(a, b, *args, **kwargs):
    print(f'a = {a}')
    print(f'b = {b}')
    print(f'args = {args}')
    print(f'kwargs = {kwargs}')

f(1, 2, 'foo', 'bar', x=100, y=200)
# Salida:
# a = 1
# b = 2
# args = ('foo', 'bar')
# kwargs = {'x': 100, 'y': 200}
````

## Función con multiples variables

Vamos a calcular el valor total de una consultoria por horas donde defines:
- Los honorarios por hora, hourly_wage
- El impuesto, tax
- La cantidad de horas, num_hours


````python
def get_pay_with_more_inputs(num_hours, hourly_wage, tax):
    # Pre-tax pay
    pay_pretax = num_hours * hourly_wage
    # After-tax pay
    pay_aftertax = pay_pretax * (1 - tax)
    return pay_aftertax
````
