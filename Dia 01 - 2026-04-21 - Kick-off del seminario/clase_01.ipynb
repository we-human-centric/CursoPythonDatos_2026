Eso que ves es lo esencial: variables, datos, un print que habla. Mañana entendemos *cómo* funciona. Hoy es solo: "esto corre".

Detalle importante: en Python los nombres de variables usan `snake_case` (palabras en minúscula, separadas por guion bajo). Es convención; todos la usan. Mejor seguirla desde el inicio.

````python
# Guiado: calcula el promedio de tres números
num1 = 10
num2 = 20
num3 = 30

promedio = (num1 + num2 + num3) / 3
print(f"El promedio es: {promedio}")

assert promedio == 20.0
print("✓ Correcto")
````

Viste ese `assert`? Es una línea que dice: "si esto no es verdad, grita un error". Los usamos para validar ejercicios. Si el assert pasa sin ruido, estás bien.

````python
# Ahora con dinero. Tres ventas del día de un vendedor.
venta1 = 150.50
venta2 = 200.00
venta3 = 75.25

total_ventas = venta1 + venta2 + venta3
comision = total_ventas * 0.10  # 10% de comisión

print(f"Total de ventas: ${total_ventas:.2f}")
print(f"Tu comisión (10%): ${comision:.2f}")

assert abs(total_ventas - 425.75) < 0.01
print("✓ Cálculo correcto")
````

Ves el `.2f`? Formato que dice: "muestra dos decimales". Con dinero, siempre quieres eso.

Y ese `abs(...) < 0.01` es un truco para evitar problemas raros de precisión con decimales. Detalle que veremos.

````python
# Ejercicio aplicado:
# Un cuadrado de lado 5. Calcula el área.
# Área = lado × lado

lado = 5
area = lado * lado

print(f"Lado: {lado}")
print(f"Área: {area}")

assert area == 25
print("✓ Ejercicio aplicado completado")
````

````python
# Ejercicio desafío (sin andamiaje):
# Un cliente compra: A=$25.99 B=$10.50 C=$45.00
# Calcula: subtotal, IVA (12%), total a pagar

precio_a = 25.99
precio_b = 10.50
precio_c = 45.00

subtotal = precio_a + precio_b + precio_c
iva = subtotal * 0.12
total = subtotal + iva

print(f"Subtotal: ${subtotal:.2f}")
print(f"IVA (12%): ${iva:.2f}")
print(f"Total a pagar: ${total:.2f}")

assert abs(total - 88.7888) < 0.01
print("✓ Ejercicio desafío completado")
````
