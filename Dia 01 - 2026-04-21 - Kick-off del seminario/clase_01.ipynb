## 1. ¿Qué es Python y por qué lo usamos?<img src="https://www.python.org/static/community_logos/python-logo-generic.svg" width="220" alt="Python logo">Python es un lenguaje de programación de **alto nivel**, **interpretado**, **tipado dinámicamente** y **multiparadigma**, creado por **Guido van Rossum** y publicado por primera vez en 1991. Su nombre es un homenaje al grupo británico de comedia Monty Python, y la filosofía del lenguaje queda resumida en *El Zen de Python* (PEP 20), que puedes leer ejecutando `import this` en cualquier intérprete.En el ecosistema de datos, Python se impuso porque combina tres cosas que otros lenguajes no ofrecen juntas: una sintaxis que lee casi como pseudocódigo, una comunidad enorme que ha construido librerías científicas (NumPy, pandas, scikit-learn, matplotlib) y una capacidad de pegamento que permite integrar trabajo hecho en C, Fortran o bases de datos sin fricción. En las encuestas anuales de Stack Overflow y Kaggle, Python es desde 2018 el lenguaje más usado en análisis de datos y aprendizaje automático.Para nuestro seminario, Python es el lenguaje común porque:- **Tiene baja curva de entrada**: en la primera hora ya se pueden escribir programas útiles.- **Es el estándar profesional en datos**: lo que aprendan se usa directamente en la industria.- **Se integra bien con todo lo demás**: bases de datos SQL, APIs web, dashboards y reportes.

## 4. ¿Qué es Jupyter y por qué usamos cuadernos?<img src="https://jupyter.org/assets/homepage/main-logo.svg" width="160" alt="Jupyter logo">Un **cuaderno Jupyter** (`.ipynb`) es un documento interactivo que mezcla **celdas de texto en Markdown** con **celdas de código ejecutable**. La idea viene de la *programación literata* de Donald Knuth (1984): que el código y su explicación vivan en el mismo documento.Cada cuaderno corre sobre un **kernel** (un proceso de Python en nuestro caso). Cuando ejecutas una celda, el kernel recuerda todo lo anterior: variables, funciones, imports. Por eso el orden en que ejecutas las celdas importa — no es lo mismo leer el cuaderno de arriba a abajo que ejecutar una celda del medio primero.Jupyter es ideal para análisis exploratorio, prototipos y enseñanza. Para productos finales (APIs, pipelines en producción) se suele pasar a archivos `.py` normales.

### Atajos útiles en Jupyter| Atajo | Acción ||---|---|| `Shift` + `Enter` | Ejecuta la celda y pasa a la siguiente || `Ctrl/Cmd` + `Enter` | Ejecuta la celda y se queda || `A` / `B` (modo comando) | Inserta celda arriba / abajo || `M` / `Y` (modo comando) | Celda a Markdown / a Código || `DD` (modo comando) | Borra la celda |El *modo comando* se activa con `Esc`; el *modo edición* con `Enter`.

## 7. Ciclo de vida de un proyecto analítico (CRISP-DM)

CRISP-DM (*Cross-Industry Standard Process for Data Mining*) es la metodología más citada para proyectos de datos desde los años 2000. Propone 6 fases iterativas:<img src="https://upload.wikimedia.org/wikipedia/commons/b/b9/CRISP-DM_Process_Diagram.png" width="420" alt="CRISP-DM diagram">1. **Entendimiento del negocio** — ¿Qué problema queremos resolver?2. **Entendimiento de los datos** — ¿Qué tenemos? ¿De qué calidad?3. **Preparación** — Limpieza, integración, transformación.4. **Modelado** — EDA, estadística, modelos.5. **Evaluación** — ¿Responde a la pregunta de negocio?6. **Despliegue** — Producto analítico (dashboard, API, reporte).El seminario atraviesa las 6 fases: llegamos hasta el *despliegue* con un dashboard Streamlit o una API FastAPI. Las flechas de retorno del diagrama son importantes: en la práctica siempre se vuelve atrás (más de una vez).
