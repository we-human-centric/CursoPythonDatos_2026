## ⏱️ Setup del entorno

Primero confirmemos que el entorno está listo. En Jupyter (en Google Colab o localmente) todo ya funciona. Si estás en tu máquina local, ejecuta estas celdas para verificar versiones.

⏱️ ~5 min · Verificación automática

````python
# --- Verificación de instalación -----------------------------------------------
import sys
import subprocess

print(f"Python {sys.version}")
print(f"Ejecutable: {sys.executable}")

# Verificar pandas y numpy
try:
    import pandas as pd
    import numpy as np
    import matplotlib.pyplot as plt
    import seaborn as sns
    print(f"\nLibrerías instaladas:")
    print(f"  pandas {pd.__version__}")
    print(f"  numpy {np.__version__}")
    print(f"  matplotlib {plt.__version__}")
    print(f"  seaborn {sns.__version__}")
    print("\n✅ Setup completo: ambiente científico listo para analítica.")
except ImportError as e:
    print(f"⚠️ Falta una librería: {e}")
    print("Ejecuta en tu terminal: pip install pandas numpy matplotlib seaborn")
````

## 🟢 Ejercicio 1 · Verificador automático de versiones

⏱️ ~3 min · Verificación de Bloom: **Recordar** y **Entender**.

**Instrucción:** La siguiente celda valida que todas las librerías necesarias están presentes. Si aparece `✅ Todas las librerías necesarias están presentes`, has pasado.

````python
# 🟢 Validación automática
required_libraries = ['pandas', 'numpy', 'matplotlib', 'seaborn']
missing = []

for lib in required_libraries:
    try:
        __import__(lib)
    except ImportError:
        missing.append(lib)

if not missing:
    print("✅ Todas las librerías necesarias están presentes.")
    print("Puedes continuar con el seminario.")
else:
    print(f"⚠️ Faltan: {', '.join(missing)}")
    print("Instálalas con: pip install " + " ".join(missing))
````

## 🔴 Ejercicio 3 · Git: inicializar repo, agregar archivo y hacer primer commit

⏱️ ~10 min · Verificación de Bloom: **Analizar** y **Crear**.

**Contexto:** Git es tu herramienta de versionado. Aquí harás tu primer commit.

**Instrucción:**
1. En tu terminal, entra a la carpeta del seminario: `cd seminario-eda`
2. Inicializa Git: `git init`
3. Verifica que funciona: `git status`
4. Crea un archivo `hola.py` con:
   ```python
   print("Hola, seminario EDA 2026")
   ```
5. Agrega el archivo: `git add hola.py`
6. Haz tu primer commit: `git commit -m "primer commit: script de bienvenida"`
7. Verifica el historial: `git log --oneline`

**Output esperado:** Un log que muestra tu commit.

> 📌 **Concepto clave:** Git registra la historia de cambios. Cada `commit` es una foto de tu proyecto en un momento específico. Estas fotos son esenciales para trabajar en equipo sin pisar cambios.
