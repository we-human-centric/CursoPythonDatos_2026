Tipos primitivos. Cada dato en Python tiene un tipo: es un número entero, un decimal, texto o un booleano (verdadero/falso). Python infiere el tipo automáticamente, pero es bueno saberlo.

```python
# Tipos primitivos
edad = 25              # int
altura = 1.75          # float
nombre = "Juan"        # str
es_estudiante = True   # bool

print(f"Nombre: {nombre}, tipo: {type(nombre).__name__}")
print(f"Edad: {edad}, tipo: {type(edad).__name__}")
print(f"Altura: {altura}, tipo: {type(altura).__name__}")
print(f"Es estudiante: {es_estudiante}, tipo: {type(es_estudiante).__name__}")
```

La función `type()` te dice qué es algo. El `.__name__` extrae solo el nombre (sino te devuelve `<class 'int'>`, que es más información de la necesaria).

```python
# Operadores aritméticos
a = 10
b = 3

suma = a + b
resta = a - b
mult = a * b
div = a / b          # División (devuelve float)
div_entera = a // b  # División entera (devuelve int)
modulo = a % b       # Residuo
potencia = a ** b    # Elevado a la potencia

print(f"Suma: {suma}")
print(f"Resta: {resta}")
print(f"Multiplicación: {mult}")
print(f"División: {div:.2f}")
print(f"División entera: {div_entera}")
print(f"Módulo (residuo): {modulo}")
print(f"Potencia: {potencia}")
```

Nota la diferencia: `/` devuelve float (3.333...), `//` devuelve entero (3). Y `%` te da el residuo (10 dividido 3 es 3 con residuo 1).

```python
# Operadores de comparación (devuelven True/False)
x = 5
y = 8

print(f"{x} == {y}: {x == y}")    # Igualdad
print(f"{x} != {y}: {x != y}")    # Desigualdad
print(f"{x} < {y}: {x < y}")      # Menor que
print(f"{x} > {y}: {x > y}")      # Mayor que
print(f"{x} <= {y}: {x <= y}")    # Menor o igual
print(f"{x} >= {y}: {x >= y}")    # Mayor o igual
```

Importante: `==` es comparación (¿son iguales?), `=` es asignación (asígna este valor). Primera regla de debug en Python: "¿por qué mi if no funciona?". Respuesta: "escribiste `=` en lugar de `==`".

```python
# Strings: slicing y métodos
mensaje = "Hola, Seminario EDA"

# Slicing: extrae partes del string
primeras_4 = mensaje[0:4]      # Posiciones 0, 1, 2, 3 (no 4)
ultimas_3 = mensaje[-3:]       # Últimas 3 letras
cada_2 = mensaje[::2]          # Cada 2 caracteres

print(f"Mensaje completo: {mensaje}")
print(f"Primeras 4: {primeras_4}")
print(f"Últimas 3: {ultimas_3}")
print(f"Cada 2 caracteres: {cada_2}")

# Métodos de string
print(f"\nMayúsculas: {mensaje.upper()}")
print(f"Minúsculas: {mensaje.lower()}")
print(f"¿Comienza con 'Hola'? {mensaje.startswith('Hola')}")
print(f"¿Termina con 'EDA'? {mensaje.endswith('EDA')}")
print(f"Reemplazar 'Seminario' por 'Curso': {mensaje.replace('Seminario', 'Curso')}")
```

Slicing es poderoso: `[inicio:fin:paso]`. Importante: `fin` no se incluye. Índices negativos cuentan desde el final. Y los métodos como `.upper()`, `.lower()`, `.replace()` vuelven la vida mucho más fácil.
