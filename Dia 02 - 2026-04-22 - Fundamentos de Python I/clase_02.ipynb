## 1. Variables y tipado dinámico

Una **variable** es un nombre que apunta a un objeto en memoria. En Python se crean al asignar un valor con `=`, sin necesidad de declarar el tipo:

```python
precio = 30        # entero
producto = "pan"   # string
```

Python es de **tipado dinámico**: una misma variable puede apuntar a objetos de tipos distintos en distintos momentos. Esto lo hace flexible pero obliga a ser cuidadoso. El tipo lo lleva el *objeto*, no la *variable*.

**Reglas de nombres** (PEP 8):
- Solo letras, dígitos y `_`. No pueden empezar por dígito.
- Se distinguen mayúsculas y minúsculas (`precio` ≠ `Precio`).
- Convención: `snake_case` para variables y funciones; `PascalCase` para clases; `MAYUSCULAS` para constantes.
- Nombres descriptivos: `precio_unitario` > `p1`.

## 2. Tipos primitivos en Python

Los **tipos primitivos** son los bloques básicos. Los que usaremos constantemente:

| Tipo | Ejemplo | Uso |
|---|---|---|
| `int` | `42` | Enteros sin tamaño máximo práctico |
| `float` | `3.14` | Números con decimales (IEEE 754) |
| `bool` | `True`, `False` | Verdadero / falso |
| `str` | `"hola"` | Cadenas de texto (Unicode) |
| `NoneType` | `None` | Ausencia de valor |

**Detalle sobre floats**: `0.1 + 0.2` en Python no es exactamente `0.3` sino `0.30000000000000004`. Esto no es un bug: es una limitación de la representación binaria de decimales. Para dinero se usa el módulo `decimal`; para comparar floats no se usa `==` sino `math.isclose()`.

**Tipado opcional**: desde Python 3.5 existen *type hints* que permiten anotar el tipo esperado. No son obligatorios ni verificados en ejecución, pero ayudan a herramientas como `mypy` y a los IDEs.

# Tipos de datos en Python

* Numéricos: enteros `int`, decimales-punto flotante `float` y complejos `complex`

* Cadenas de Caracteres: datos estructurados a partir de `str` (texto y palabras). No hay diferencia en el uso de comillas simples y dobles. La comillas triples se emplean para escribir múltiples líneas (párrafos)

```type(x)``` es una función de Python que nos permite saber por consola el tipo de dato que es x.

## 3. Operadores aritméticos

Python incluye los operadores habituales y dos que conviene memorizar:

| Operador | Acción | Ejemplo |
|---|---|---|
| `+ - * /` | Suma, resta, multiplicación, división | `7 / 2 == 3.5` |
| `//` | División entera (sin decimales) | `7 // 2 == 3` |
| `%` | Módulo (residuo) | `7 % 2 == 1` |
| `**` | Potencia | `2 ** 10 == 1024` |

El módulo (`%`) es útil para pruebas de paridad (`n % 2 == 0` significa "par") y para operaciones cíclicas.

**Precedencia**: como en matemáticas, `** `tiene más precedencia que `*` y `/`, que a su vez tienen más precedencia que `+` y `-`. En caso de duda, pon paréntesis: hacen el código legible.

## 4. Operadores relacionales y lógicos

Los **operadores relacionales** (`==`, `!=`, `<`, `<=`, `>`, `>=`) comparan dos valores y devuelven un booleano. Ojo: `=` asigna, `==` compara. Confundirlos es un error clásico.

Los **operadores lógicos** combinan booleanos:

- `and` — verdadero si ambos lo son (conjunción).
- `or` — verdadero si al menos uno lo es (disyunción).
- `not` — niega el valor.

Python evalúa en **corto circuito**: en `a and b`, si `a` es falso, `b` ni siquiera se evalúa. Esto se explota para escribir código defensivo como `if obj is not None and obj.valor > 0:`.

**Operadores de pertenencia** (`in`, `not in`) permiten preguntar si un elemento está dentro de una colección: `"a" in "hola"` es `True`.

# Operadores Relacionales y lógicos
* Operadores **RELACIONALES**: Comparan valores y devuelven un booleano (```True``` o ```False```) según corresponda el resultado de la evaluación.

````python
a = 9 # Almacena el valor 9 en a
a == 8 # Confirmame si a es igual a 8
````

````python
a=4
b=18
print("mayor que:", a > b)
print("menor que:", a < b)
````

````python
a, b = 1, 5
print("mayor o igual que:", b >= a)    # 5 >= 1
print("menor o igual que:", -1 <= a)   # -1 <= 1
````

````python
# ¿Es verdadera esta expresión?
5 > 3 > 1
````

````python
# ¿Es verdadera esta expresión?
5 > 3 < 4 == 3 + 1
````

# Operadores **LÓGICOS**:

Permiten construir expresiones lógicas. El resultado es un booleano.
>NOTA: Revisa [tablas de verdad](https://es.wikipedia.org/wiki/Tabla_de_verdad) para las operaciones de conjunción, disyunción y negación.


````python
# CONJUNCION (and)
a = 4
a >= 1 and a <= 5   #(4>=1)  and  (4<=5)  = True
````

````python
# DISYUNCION (or)
a = 7
a >= 6 or a == 10   #(7>=6) or (a==10) = True
````

````python
# NEGACION (not)
a = False
not a
````

## ⚠️ Prioridad de Operadores:

**Nota** Cuando se utilizan varios operadores en una sola expresión, la precedencia de los operadores determina qué partes de la expresión se evalúan y en qué orden. Los operadores con mayor precedencia se evalúan primero. Los operadores con la misma precedencia se evalúan de izquierda a derecha.

Orden de los operadores:
- () parentheses, for grouping
- ** exponent
- *, / multiplication and division
- +, - addition and subtraction
- ==, !=, <, <=, >, >= comparisons

En resumen:

1.   Parentesis
2.   Potencia
3.  Multiplicación y división
4.  Suma y resta
5. Comparaciones


````python
y = -1 + 2 * 0 - (24 / 2 ** 3)

print(y)
````

## 5. Strings · representación de texto

Las cadenas (`str`) son secuencias **inmutables** de caracteres Unicode. Eso significa que una vez creadas no se modifican: las operaciones que "cambian" un string en realidad devuelven uno nuevo.

Las cadenas se indexan desde `0` y soportan **slicing** (`s[a:b:paso]`). Los índices negativos cuentan desde el final (`s[-1]` es el último carácter).

### f-strings (desde Python 3.6)

La forma moderna de componer texto con variables:

```python
nombre = "Ana"
edad = 30
print(f"{nombre} tiene {edad} años")
```

Dentro de las llaves se puede poner cualquier expresión Python, no solo nombres: `f"{2 + 2}"`, `f"{precio * 1.12:.2f}"`, `f"{fecha:%Y-%m-%d}"`. Las f-strings son legibles, rápidas y permiten formateo (decimales, padding, fechas).

# **Strings en Python**




**Cadenas:**
El tipo str en Python se refiere a las cadenas de caracteres, que son secuencias de caracteres Unicode. Las cadenas en Python pueden contener letras, números, símbolos y caracteres especiales, y se utilizan para representar texto.

**Índices y slicing o rebanadas:** Los strings se comportan como listas, puedes hacerles slicing. Puedes acceder a caracteres individuales en una cadena utilizando índices y puedes obtener subcadenas utilizando slicing. Esto veremos despues en listas.

````python
cadena = "Hola"
cadena[0]
````

````python
cadena[3]
````

**Inmutable:** Las cadenas en Python son objetos inmutables, lo que significa que una vez que se crea una cadena, no se puede modificar. Si necesitas modificar una cadena, debes crear una nueva.


```python
cadena = "Hola"
# Esto generaría un error, ya que las cadenas son inmutables
cadena[0] = 'h'
```



**Concatenación:** Puedes concatenar cadenas utilizando el operador +.

````python
saludo = "Hola"
nombre = "Mundo"
saludo_completo = saludo + " " + nombre
print(saludo_completo)
````

**Formato de cadena:** Python ofrece varias formas de formatear cadenas, incluyendo el método format() y las f-strings (cadenas f)

````python
nombre = "Juan"
edad = 30
print("Hola, mi nombre es {} y tengo {} años.".format(nombre, edad))
````

**Cadena multilínea:** Puedes crear cadenas multilínea utilizando comillas triples (""" o ''').

````python
poema = """
En un lugar de la Mancha,
de cuyo nombre no quiero acordarme,
no ha mucho tiempo que vivía un hidalgo...
"""
print(poema)
````

**Las f-strings** son una característica introducida en Python 3.6 que te permite insertar valores de variables directamente dentro de una cadena de una manera muy legible y concisa.

````python
nombre = "Mundo"
saludo = f"Hola {nombre}!"
print(saludo)
````

````python
# Puedes incluir cualquier expresión Python válida dentro de las llaves {}.
numero = 42
saludo = f"El doble de {numero} es {numero * 2}."
print(saludo)
````

````python
# Puedes incluir expresiones matemáticas simples
a = 10
b = 5
resultado = f"La suma de {a} y {b} es {a + b}, y su resta es {a - b}."
print(resultado)
````

````python
# Puedes redondear números dentro de una f-string utilizando la función round().

pi = 3.14159265358979323846
resultado = f"El valor de pi redondeado a dos decimales es {round(pi, 2)}."
print(resultado)
````

````python
# Puedes utilizar cadenas de formato dentro de f-strings

precio = 99.9595
resultado = f"El precio es ${precio:.2f}."
print(resultado)
````
