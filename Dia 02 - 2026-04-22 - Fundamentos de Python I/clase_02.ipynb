## 📌 Concepto clave · Variables y tipos primitivos

Una **variable** es un nombre para un valor. Los **tipos primitivos** son los bloques básicos:
- `int` → enteros (42, -5, 0)
- `float` → decimales (3.14, -0.5)
- `str` → texto ("hola", 'mundo')
- `bool` → verdadero/falso (True, False)
- `None` → ausencia de valor

Python es **tipado dinámico**: una variable puede cambiar de tipo.

### ⏱️ ~5 min · Conocer tipos con `type()`

````python
# Ejemplos de tipos primitivos
edad = 25
altura = 1.75
nombre = "Juan"
es_estudiante = True
valor_nulo = None

# Cómo preguntar el tipo
print(f"{edad} es {type(edad)}")
print(f"{altura} es {type(altura)}")
print(f"{nombre} es {type(nombre)}")
print(f"{es_estudiante} es {type(es_estudiante)}")
print(f"{valor_nulo} es {type(valor_nulo)}")
````

## 📌 Concepto clave · Operadores aritméticos, relacionales y lógicos

| Tipo | Operadores | Ejemplo |
|---|---|---|
| Aritméticos | `+ - * / // % **` | `10 // 3` → 3 |
| Relacionales | `== != < > <= >=` | `5 > 3` → True |
| Lógicos | `and or not` | `(5 > 3) and (2 < 4)` → True |

**Detalle importante:** `=` asigna, `==` compara. Confundirlos es un error clásico.

### ⏱️ ~7 min · Demo: operadores en acción

````python
# Aritméticos
print("Aritméticos:")
print(f"7 + 3 = {7 + 3}")
print(f"7 / 2 = {7 / 2}")
print(f"7 // 2 = {7 // 2}")
print(f"7 % 2 = {7 % 2}")
print(f"2 ** 10 = {2 ** 10}")

# Relacionales
print("\nRelacionales:")
print(f"7 == 7 es {7 == 7}")
print(f"7 != 8 es {7 != 8}")
print(f"7 > 5 es {7 > 5}")

# Lógicos
print("\nLógicos:")
print(f"(5 > 3) and (2 < 4) es {(5 > 3) and (2 < 4)}")
print(f"(5 > 3) or (2 > 4) es {(5 > 3) or (2 > 4)}")
print(f"not True es {not True}")
````

## 📌 Concepto clave · Strings y f-strings

Las cadenas (`str`) son secuencias de caracteres. Puedes:
- **Concatenar:** `"Hola" + " " + "mundo"`
- **Indexar:** `"hola"[0]` → `"h"`
- **Slicear:** `"hola"[1:3]` → `"ol"`
- **Usar f-strings:** `f"{variable} texto"` (la forma moderna y recomendada)

**F-strings permiten:**
- Insertar variables: `f"Hola {nombre}"`
- Expresiones: `f"Total: {precio * 1.12:.2f}"`
- Formateo: 2 decimales, padding, etc.

### ⏱️ ~5 min · Demo: strings y f-strings

````python
# Indexación
palabra = "hola"
print(f"Primera letra de '{palabra}': {palabra[0]}")
print(f"Última letra: {palabra[-1]}")

# Slicing
print(f"Letras 1 a 3: {palabra[1:3]}")

# F-strings con expresiones
temperatura = 25.456
print(f"Temperatura redondeada: {temperatura:.1f}°C")

# Métodos de string
frase = "  Python es genial  "
print(f"Original: '{frase}'")
print(f"Sin espacios: '{frase.strip()}'")
print(f"Mayúsculas: '{frase.upper()}'")
````
