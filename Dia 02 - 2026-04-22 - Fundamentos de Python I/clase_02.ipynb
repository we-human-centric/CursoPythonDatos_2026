```python
import random
import numpy as np
import pandas as pd

SEED = 42
random.seed(SEED)
np.random.seed(SEED)

print("Setup completo ✓")
```

Tipos primitivos. Cada dato en Python tiene un tipo: es un número entero, un decimal, texto o un booleano (verdadero/falso). Python infiere el tipo automáticamente, pero es bueno saberlo.

```python
# Tipos primitivos
edad = 25              # int
altura = 1.75          # float
nombre = "Juan"        # str
es_estudiante = True   # bool

print(f"Nombre: {nombre}, tipo: {type(nombre).__name__}")
print(f"Edad: {edad}, tipo: {type(edad).__name__}")
print(f"Altura: {altura}, tipo: {type(altura).__name__}")
print(f"Es estudiante: {es_estudiante}, tipo: {type(es_estudiante).__name__}")
```

La función `type()` te dice qué es algo. El `.__name__` extrae solo el nombre (sino te devuelve `<class 'int'>`, que es más información de la necesaria).

```python
# Operadores aritméticos
a = 10
b = 3

suma = a + b
resta = a - b
mult = a * b
div = a / b          # División (devuelve float)
div_entera = a // b  # División entera (devuelve int)
modulo = a % b       # Residuo
potencia = a ** b    # Elevado a la potencia

print(f"Suma: {suma}")
print(f"Resta: {resta}")
print(f"Multiplicación: {mult}")
print(f"División: {div:.2f}")
print(f"División entera: {div_entera}")
print(f"Módulo (residuo): {modulo}")
print(f"Potencia: {potencia}")
```

Nota la diferencia: `/` devuelve float (3.333...), `//` devuelve entero (3). Y `%` te da el residuo (10 dividido 3 es 3 con residuo 1).

```python
# Operadores de comparación (devuelven True/False)
x = 5
y = 8

print(f"{x} == {y}: {x == y}")    # Igualdad
print(f"{x} != {y}: {x != y}")    # Desigualdad
print(f"{x} < {y}: {x < y}")      # Menor que
print(f"{x} > {y}: {x > y}")      # Mayor que
print(f"{x} <= {y}: {x <= y}")    # Menor o igual
print(f"{x} >= {y}: {x >= y}")    # Mayor o igual
```

Importante: `==` es comparación (¿son iguales?), `=` es asignación (asígna este valor). Primera regla de debug en Python: "¿por qué mi if no funciona?". Respuesta: "escribiste `=` en lugar de `==`".

```python
# Guiado: convierte tipos
numero_texto = "42"
numero_entero = int(numero_texto)     # Texto a int
numero_decimal = float(numero_entero) # Int a float
back_to_texto = str(numero_decimal)   # Float a texto

print(f"Original (texto): {numero_texto}, tipo: {type(numero_texto).__name__}")
print(f"Convertido a int: {numero_entero}, tipo: {type(numero_entero).__name__}")
print(f"Convertido a float: {numero_decimal}, tipo: {type(numero_decimal).__name__}")
print(f"De vuelta a texto: {back_to_texto}, tipo: {type(back_to_texto).__name__}")

assert numero_entero == 42
assert numero_decimal == 42.0
print("✓ Conversiones correctas")
```

Conversión de tipos. A veces necesitas convertir: `input()` devuelve texto, pero quieres un número. O quieres mostrar un número como texto. Las funciones `int()`, `float()`, `str()` lo hacen.

```python
# Strings: slicing y métodos
mensaje = "Hola, Seminario EDA"

# Slicing: extrae partes del string
primeras_4 = mensaje[0:4]      # Posiciones 0, 1, 2, 3 (no 4)
ultimas_3 = mensaje[-3:]       # Últimas 3 letras
cada_2 = mensaje[::2]          # Cada 2 caracteres

print(f"Mensaje completo: {mensaje}")
print(f"Primeras 4: {primeras_4}")
print(f"Últimas 3: {ultimas_3}")
print(f"Cada 2 caracteres: {cada_2}")

# Métodos de string
print(f"\nMayúsculas: {mensaje.upper()}")
print(f"Minúsculas: {mensaje.lower()}")
print(f"¿Comienza con 'Hola'? {mensaje.startswith('Hola')}")
print(f"¿Termina con 'EDA'? {mensaje.endswith('EDA')}")
print(f"Reemplazar 'Seminario' por 'Curso': {mensaje.replace('Seminario', 'Curso')}")
```

Slicing es poderoso: `[inicio:fin:paso]`. Importante: `fin` no se incluye. Índices negativos cuentan desde el final. Y los métodos como `.upper()`, `.lower()`, `.replace()` vuelven la vida mucho más fácil.

```python
# F-strings: la forma moderna de formatear
nombre_usuario = "María"
edad_usuario = 21
promedio_usuario = 4.5

# Vieja forma (evítala)
# mensaje_viejo = "Hola, " + nombre_usuario + ". Tienes " + str(edad_usuario) + " años."

# Nueva forma (f-strings)
mensaje = f"Hola, {nombre_usuario}. Tienes {edad_usuario} años."
calificacion = f"Promedio: {promedio_usuario:.1f}"  # 1 decimal
saludo_formateado = f"Mensaje: {mensaje.upper()}"

print(mensaje)
print(calificacion)
print(saludo_formateado)

assert "María" in mensaje
assert "4.5" in calificacion
print("✓ F-strings funcionan")
```

F-strings (f"...") son la forma moderna. Dentro de las llaves `{}` puedes poner variables, expresiones, y formato (como `.1f` para 1 decimal). Mucho más legible que concatenar con `+`.

```python
# Aplicado: calculadora simple
# Lee dos números, suma, resta, mult, div

num1 = 15.5
num2 = 3.2

suma = num1 + num2
resta = num1 - num2
multi = num1 * num2
divi = num1 / num2

resultado = f"""
Operaciones con {num1} y {num2}:
  Suma: {suma:.2f}
  Resta: {resta:.2f}
  Multiplicación: {multi:.2f}
  División: {divi:.2f}
"""

print(resultado)

assert abs(suma - 18.7) < 0.01
assert abs(divi - 4.84) < 0.01
print("✓ Calculadora funcionó")
```

Las f-strings con triple comilla (`"""..."""") permiten strings multilinea. Útil para mensajes largos o salidas formateadas.

```python
# Desafío: validador simple de email
# Un email "válido" tiene exactamente un @, y al menos un . después del @

email = "juan.perez@example.com"

cuenta_arroba = email.count('@')
posicion_arroba = email.find('@')
posicion_punto = email.find('.', posicion_arroba)  # Busca punto después del @

es_valido = (cuenta_arroba == 1) and (posicion_punto > posicion_arroba)

print(f"Email: {email}")
print(f"¿Válido? {es_valido}")

assert es_valido == True
print("✓ Validación completada")
```
