## Introducción

Hoy transformamos el caso de estudio en un proyecto analítico real. Definimos el problema de negocio, la pregunta analítica, las hipótesis iniciales y la estructura del repositorio.

## 1. Del caso de negocio a la pregunta analítica

Un **caso de negocio** es un problema expresado en lenguaje de negocio. Un **proyecto analítico** es ese problema traducido a algo que se puede responder con datos. El paso de uno al otro es el que hace fracasar la mayoría de proyectos cuando se salta.

Para hacer bien este paso conviene escribir explícitamente tres cosas:

1. **Problema**: qué está pasando y a quién le duele. Una frase.
2. **Pregunta analítica**: una pregunta específica que se pueda responder con los datos disponibles. Evitar preguntas tan abiertas que no se puedan responder nunca ("¿cómo mejoramos el negocio?").
3. **Decisión**: qué hará el negocio con la respuesta. Si nadie va a actuar sobre los resultados, el proyecto no es prioritario.

**Ejemplo (caso Spotify)**:
- Problema: no sabemos qué canciones tienen más chance de llegar al top 100 semanal.
- Pregunta: ¿qué atributos musicales (valence, energy, tempo, loudness) correlacionan con estar en el top 100?
- Decisión: priorizar promoción de canciones cuyos atributos coincidan con ese perfil.

## 2. Niveles de análisis (Gartner)

--

<img src="https://www.ixpantia.com/hubfs/Imported_Blog_Media/modelo_gartner-2-1.png" >

Un mismo dataset puede responder preguntas muy distintas. Gartner popularizó una escalera de 4 niveles:

1. **Descriptivo** — ¿Qué pasó? (conteos, promedios, series de tiempo)
2. **Diagnóstico** — ¿Por qué pasó? (segmentación, cohortes, correlación)
3. **Predictivo** — ¿Qué va a pasar? (modelos estadísticos / ML)
4. **Prescriptivo** — ¿Qué deberíamos hacer? (optimización, simulación)

Cada nivel requiere más datos y más técnica que el anterior. Un proyecto de 4 semanas es realista si se queda entre **descriptivo y predictivo simple**. No intenten el prescriptivo al primer intento.

## 3. Formulación de hipótesis

Una **hipótesis** es una afirmación verificable con los datos. Las buenas hipótesis:

- Son **específicas**: mencionan variables concretas.
- Son **falsables**: los datos podrían contradecirlas.
- Aportan valor: si se confirman o refutan, se aprende algo útil.

**Ejemplos de hipótesis buenas vs malas** (caso Titanic):

| Mala | Buena |
|---|---|
| "Las mujeres sobrevivieron más" (ya es obvio, sin magnitud) | "La tasa de supervivencia de mujeres fue al menos 3 veces la de hombres" |
| "La edad importaba" (ambiguo) | "Los pasajeros menores de 15 años tuvieron mayor tasa de supervivencia que los adultos" |

Para cada proyecto pediremos **3 a 5 hipótesis** documentadas desde la primera semana. Estas guiarán el EDA.

## 4. Estructura del repositorio

Una convención sólida:

```
mi-proyecto-eda/
├── data/
│   ├── raw/            # datos originales (inmutables)
│   ├── interim/        # datos en transformación
│   └── processed/      # datos listos para modelar
├── notebooks/
│   ├── 01_eda.ipynb
│   ├── 02_features.ipynb
│   └── 03_modelo.ipynb
├── src/
│   ├── __init__.py
│   ├── io.py           # lectura/escritura
│   ├── features.py     # transformaciones
│   └── modelos.py
├── tests/
├── app/                # Streamlit / FastAPI
├── reports/            # gráficos y documentos finales
├── requirements.txt
├── .gitignore          # ¡no subir data/raw ni .venv!
└── README.md
```

**Reglas de oro**:
- Los datos en `data/raw/` **nunca** se editan a mano.
- Los notebooks se numeran para que se lean en orden.
- El código reusable vive en `src/`, no dentro de los notebooks.
- `requirements.txt` congela las versiones de las librerías.
