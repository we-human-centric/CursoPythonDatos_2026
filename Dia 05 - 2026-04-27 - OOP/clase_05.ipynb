## 1. ¿Qué es la Programación Orientada a Objetos?

En OOP, el código se organiza alrededor de **objetos** que combinan **datos** (atributos) y **comportamientos** (métodos). Es como el mundo real: un auto tiene marca, color y precio (datos), y puede arrancar, detenerse y acelerar (comportamientos).

Los tres conceptos fundamentales que verás hoy:

| Concepto | ¿Qué es? | Analogía |
|---|---|---|
| **Clase** | El plano o molde | El diseño arquitectónico de una casa |
| **Objeto** | Una instancia del plano | Una casa específica construida con ese plano |
| **Método** | Una función del objeto | Lo que la casa puede hacer (ventilarse, iluminarse) |

```
Clase (molde)         Objetos (instancias)
┌──────────────┐      ┌──────────────────┐
│  class Auto  │  →   │ auto_rojo        │
│  ─────────── │      │ auto_azul        │
│  marca       │      │ auto_verde       │
│  modelo      │      └──────────────────┘
│  arrancar()  │
└──────────────┘
```

### Módulos y Librerías: la jerarquía que ya conoces

Las librerías que usarás en el curso están construidas con clases:

- Una **clase** agrupa datos y funciones relacionadas (ej: `DataFrame` en Pandas)
- Un **módulo** agrupa varias clases relacionadas (ej: el módulo `pandas.core`)
- Una **librería** agrupa módulos que resuelven un problema específico (ej: `pandas`)

Cuando escribes `df = pd.DataFrame(data)`, estás creando un **objeto** de la clase `DataFrame`. Y cuando ejecutas `df.describe()`, estás llamando un **método** de ese objeto.

## 2. Tu primera clase: el molde básico

````python
# La clase más simple posible
class MiClase:
    """Esta es una clase simple de ejemplo."""
    atributo = "un valor de clase"

    def metodo(self):
        """Un método de instancia básico."""
        return "Este es un método de la clase"


# Crear un objeto (instancia) a partir de la clase
mi_objeto = MiClase()

print(mi_objeto.atributo)   # Accede al atributo
print(mi_objeto.metodo())   # Llama al método
print(type(mi_objeto))      # Muestra el tipo del objeto
````

## 3. El constructor `__init__`: inicializar objetos con datos propios

El método `__init__` se ejecuta automáticamente al crear cada objeto. Es donde defines los **atributos de instancia** — los datos propios de ese objeto específico (no compartidos con otros objetos de la misma clase).

````python
class Auto:
    """Representa un automóvil con marca y modelo."""

    def __init__(self, marca, modelo, anio):
        # Los atributos de instancia se definen con self.nombre
        self.marca = marca
        self.modelo = modelo
        self.anio = anio
        self.encendido = False  # Atributo con valor por defecto

    def arrancar(self):
        """Enciende el auto si estaba apagado."""
        if not self.encendido:
            self.encendido = True
            return f"{self.marca} {self.modelo}: ¡Motor encendido!"
        return f"{self.marca} {self.modelo}: Ya estaba encendido."

    def apagar(self):
        """Apaga el auto."""
        self.encendido = False
        return f"{self.marca} {self.modelo}: Motor apagado."

    def mostrar_info(self):
        """Devuelve información del auto."""
        estado = "encendido" if self.encendido else "apagado"
        return f"{self.anio} {self.marca} {self.modelo} — {estado}"


# Crear dos autos distintos con el mismo molde
auto1 = Auto("Toyota", "Corolla", 2022)
auto2 = Auto("Hyundai", "Tucson", 2023)

print(auto1.mostrar_info())
print(auto1.arrancar())
print(auto1.mostrar_info())  # Ahora está encendido
print()
print(auto2.mostrar_info())  # auto2 sigue apagado — son objetos independientes
print(auto2.arrancar())
````

Cada objeto tiene **su propio estado**. `auto1.encendido` y `auto2.encendido` son variables independientes aunque pertenezcan a la misma clase. Esto es la esencia de la orientación a objetos.

## 4. Herencia: construir sobre lo existente

La herencia permite crear **clases especializadas** a partir de una clase base. La subclase hereda todos los atributos y métodos de la clase padre, y puede agregar los suyos o modificar los existentes.

```
        ┌─────────────┐
        │  Vehiculo   │  ← Clase base (padre / superclase)
        │  marca      │
        │  modelo     │
        │  descripcion│
        └──────┬──────┘
               │
       ┌───────┼──────────┐
       ▼                  ▼
  ┌─────────┐        ┌─────────┐
  │  Coche  │        │  Bici   │  ← Subclases (hijas)
  │  puertas│        │  aro    │
  └─────────┘        └─────────┘
```

````python
class Vehiculo:
    """Clase base para cualquier tipo de vehículo."""

    def __init__(self, marca, modelo):
        self.marca = marca
        self.modelo = modelo

    def descripcion(self):
        return f"Vehículo: {self.marca} {self.modelo}"

    def tipo(self):
        return "Vehículo genérico"


class Coche(Vehiculo):  # Hereda de Vehiculo
    """Un coche tiene puertas además de marca y modelo."""

    def __init__(self, marca, modelo, puertas):
        super().__init__(marca, modelo)  # Llama al __init__ del padre
        self.puertas = puertas

    def descripcion(self):  # Sobreescribe el método del padre
        return f"{self.marca} {self.modelo} con {self.puertas} puertas"

    def tipo(self):
        return "Automóvil"


class Bici(Vehiculo):
    """Una bici tiene propósito y tamaño de aro."""

    def __init__(self, marca, modelo, proposito, aro):
        super().__init__(marca, modelo)
        self.proposito = proposito
        self.aro = aro

    def descripcion(self):
        return f"{self.marca} {self.modelo} — aro {self.aro}\" — {self.proposito}"

    def tipo(self):
        return "Bicicleta"


# Probamos la jerarquía
vehiculos = [
    Coche("Toyota", "Corolla", 4),
    Bici("Trek", "Hardtail 12", "Montaña", 29),
    Coche("Mazda", "CX-5", 4),
]

for v in vehiculos:
    print(f"[{v.tipo()}] {v.descripcion()}")
````

**`super()`** es la forma de llamar al constructor o método del padre. Sin él, tendrías que escribir el código de inicialización duplicado en cada subclase.

¿Qué tipo de objeto es `mi_bici`? Veamos:

````python
mi_bici = Bici("Trek", "Hardtail", "Montaña", 29)

print(type(mi_bici))                          # Es una Bici
print(isinstance(mi_bici, Bici))              # True: es una Bici
print(isinstance(mi_bici, Vehiculo))          # True: también es un Vehiculo
print(isinstance(mi_bici, Coche))             # False: NO es un Coche
````

## 5. Polimorfismo: misma interfaz, distinto comportamiento

Polimorfismo significa que objetos de distintas clases pueden responder al mismo mensaje (método) de forma diferente. Esto permite escribir código genérico que funciona con cualquier tipo de objeto, siempre que tenga el método esperado.

````python
class Perro:
    def __init__(self, nombre):
        self.nombre = nombre

    def sonido(self):
        return "Guau"

    def presentarse(self):
        return f"Hola, soy {self.nombre} y digo: {self.sonido()}"


class Gato:
    def __init__(self, nombre):
        self.nombre = nombre

    def sonido(self):
        return "Miau"

    def presentarse(self):
        return f"Hola, soy {self.nombre} y digo: {self.sonido()}"


class Pato:
    def __init__(self, nombre):
        self.nombre = nombre

    def sonido(self):
        return "Cuac"

    def presentarse(self):
        return f"Hola, soy {self.nombre} y digo: {self.sonido()}"


# La función no sabe qué tipo de animal recibe — solo que tiene .sonido()
def hacer_ruido(animal):
    print(f"{type(animal).__name__}: {animal.sonido()}")


animales = [Perro("Buddy"), Gato("Luna"), Pato("Donald"), Perro("Rex")]

for animal in animales:
    hacer_ruido(animal)

print()
for animal in animales:
    print(animal.presentarse())
````

## 6. Encapsulamiento: controlar el acceso a los datos

El encapsulamiento permite ocultar los detalles internos de una clase y exponer solo lo necesario. En Python se usan convenciones:

- `atributo` → público (accesible desde cualquier lugar)
- `_atributo` → privado por convención ("no lo toques desde afuera")
- `__atributo` → privado con name mangling (Python lo renombra para protegerlo)

````python
class CuentaBancaria:
    """Cuenta bancaria con saldo protegido."""

    def __init__(self, titular, saldo_inicial):
        self.titular = titular       # Público
        self.__saldo = saldo_inicial  # Privado — no accesible directamente desde afuera
        self.__historial = []         # Privado

    def depositar(self, cantidad):
        """Agrega dinero a la cuenta."""
        if cantidad <= 0:
            raise ValueError("El depósito debe ser positivo.")
        self.__saldo += cantidad
        self.__historial.append(f"Depósito: +${cantidad:.2f}")
        return f"Depositado ${cantidad:.2f}. Nuevo saldo: ${self.__saldo:.2f}"

    def retirar(self, cantidad):
        """Retira dinero si hay fondos suficientes."""
        if cantidad <= 0:
            raise ValueError("El retiro debe ser positivo.")
        if cantidad > self.__saldo:
            raise ValueError("Fondos insuficientes.")
        self.__saldo -= cantidad
        self.__historial.append(f"Retiro: -${cantidad:.2f}")
        return f"Retirado ${cantidad:.2f}. Saldo restante: ${self.__saldo:.2f}"

    def obtener_saldo(self):
        """Acceso controlado al saldo."""
        return self.__saldo

    def ver_historial(self):
        """Devuelve el historial de transacciones."""
        return self.__historial.copy()  # Copia para que no puedan modificarlo


# Uso
cuenta = CuentaBancaria("Ana García", 1000.00)

print(cuenta.depositar(500))
print(cuenta.retirar(200))
print(f"Saldo actual: ${cuenta.obtener_saldo():.2f}")
print("\nHistorial:")
for movimiento in cuenta.ver_historial():
    print(f"  {movimiento}")

# Intentar acceder al saldo directamente:
try:
    print(cuenta.__saldo)  # Esto falla
except AttributeError as e:
    print(f"\nError esperado: {e}")
````

## 7. Métodos estáticos y propiedades

### `@staticmethod`: métodos que no dependen del objeto

Un método estático no recibe `self` — no necesita ni modifica datos del objeto. Se llama en la clase directamente.

````python
class Conversor:
    """Utilidades de conversión de unidades."""

    @staticmethod
    def celsius_a_fahrenheit(c):
        return c * 9/5 + 32

    @staticmethod
    def km_a_millas(km):
        return km * 0.621371

    @staticmethod
    def usd_a_eur(usd, tasa=0.92):
        return usd * tasa


# Se llaman SIN crear un objeto — directamente en la clase
print(f"37°C = {Conversor.celsius_a_fahrenheit(37):.1f}°F")
print(f"100 km = {Conversor.km_a_millas(100):.2f} millas")
print(f"$100 USD = €{Conversor.usd_a_eur(100):.2f}")
````

### `@property`: atributos con lógica de acceso y validación

````python
class Persona:
    """Persona con nombre y edad validados."""

    def __init__(self, nombre, edad):
        self.__nombre = nombre
        self.__edad = edad

    @property
    def nombre(self):
        """Getter: accede al nombre como si fuera un atributo normal."""
        return self.__nombre

    @nombre.setter
    def nombre(self, valor):
        """Setter: valida antes de asignar."""
        if not valor or not isinstance(valor, str):
            raise ValueError("El nombre debe ser una cadena no vacía.")
        self.__nombre = valor.strip().title()  # Normaliza capitalización

    @property
    def edad(self):
        return self.__edad

    @edad.setter
    def edad(self, valor):
        if not isinstance(valor, int) or valor < 0 or valor > 130:
            raise ValueError("Edad inválida.")
        self.__edad = valor

    def __str__(self):
        return f"{self.nombre} ({self.edad} años)"


p = Persona("julián maya", 35)
print(p)              # Imprime la representación del objeto
print(p.nombre)       # Se accede como atributo, no como método

p.nombre = "  andrés torres  "  # Se normaliza automáticamente
print(p.nombre)

try:
    p.edad = -5       # Falla la validación
except ValueError as e:
    print(f"Error: {e}")
````

## 8. Métodos mágicos (Dunder methods)

Los métodos mágicos (también llamados *dunder* por el doble guión bajo) permiten que tus objetos se integren con Python de forma natural: puedes sumarlos, compararlos, imprimirlos, medirlos. Son los que hacen que `len(mi_lista)` o `print(mi_objeto)` funcionen.

````python
class Punto:
    """Punto en un espacio 2D con operaciones matemáticas."""

    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __str__(self):
        """Representación amigable para print()."""
        return f"Punto({self.x}, {self.y})"

    def __repr__(self):
        """Representación técnica (para debug y notebooks)."""
        return f"Punto(x={self.x}, y={self.y})"

    def __add__(self, otro):
        """Permite usar + entre dos puntos."""
        return Punto(self.x + otro.x, self.y + otro.y)

    def __sub__(self, otro):
        """Permite usar - entre dos puntos."""
        return Punto(self.x - otro.x, self.y - otro.y)

    def __eq__(self, otro):
        """Permite comparar con ==."""
        return self.x == otro.x and self.y == otro.y

    def distancia(self, otro):
        """Distancia euclídea entre dos puntos."""
        return ((self.x - otro.x)**2 + (self.y - otro.y)**2) ** 0.5


p1 = Punto(1, 2)
p2 = Punto(4, 6)

print(p1)                          # Usa __str__
print(repr(p1))                    # Usa __repr__
print(f"Suma: {p1 + p2}")         # Usa __add__
print(f"Resta: {p2 - p1}")        # Usa __sub__
print(f"¿Son iguales? {p1 == p2}")  # Usa __eq__
print(f"Distancia: {p1.distancia(p2):.2f}")
````

````python
# __len__ y comparaciones con personas
class Equipo:
    """Equipo de trabajo con una lista de miembros."""

    def __init__(self, nombre, miembros):
        self.nombre = nombre
        self.miembros = list(miembros)

    def __len__(self):
        """Permite usar len() con el equipo."""
        return len(self.miembros)

    def __str__(self):
        return f"Equipo '{self.nombre}' ({len(self)} miembros)"

    def __contains__(self, persona):
        """Permite usar 'in' para verificar si alguien está en el equipo."""
        return persona in self.miembros


equipo = Equipo("Data Science", ["Ana", "Luis", "Carla", "Marcos"])

print(equipo)
print(f"Tamaño: {len(equipo)}")         # Usa __len__
print(f"¿Está Ana? {'Ana' in equipo}")   # Usa __contains__
print(f"¿Está Pedro? {'Pedro' in equipo}")
````
