## 1. Data storytelling · no es solo mostrar númerosUna **presentación analítica** no es un reporte técnico ni una demo de código. Es un **relato** que lleva al público de un problema a una decisión. Los mejores analistas del mundo (Hans Rosling, Edward Tufte, Cole Nussbaumer) coinciden en tres ingredientes:1. **Contexto**: por qué este problema importa.2. **Tensión**: qué tensión o sorpresa reveló el análisis.3. **Resolución**: qué acción concreta se propone.Las **tres preguntas** que debe poder contestar cualquier persona que vea tu presentación:- ¿Cuál era el problema?- ¿Qué descubriste?- ¿Qué deberíamos hacer?Si el público no puede responder eso en 30 segundos después de tu pitch, la presentación falló.

## 2. Pirámide de Minto · estructura de comunicación<img src="https://upload.wikimedia.org/wikipedia/commons/thumb/6/62/Minto_pyramid_principle.png/640px-Minto_pyramid_principle.png" width="360" alt="Pirámide de Minto">Desarrollada en McKinsey por **Barbara Minto** en los años 60, esta técnica consiste en:1. **Empezar con la conclusión** (la punta de la pirámide).2. Apoyarla con **3 argumentos principales**.3. Cada argumento apoyado por **evidencia** (datos, gráficos).Lo contrario es el **suspense académico**: "ahora les cuento todo lo que hicimos y al final les digo qué encontramos". Evítenlo. Los ejecutivos se aburren y se van.

## 3. Estructura de 10 minutos| Minuto | Sección | Qué decir ||---|---|---|| 0:00–1:00 | **Contexto y problema** | El caso, por qué importa, a quién afecta || 1:00–2:00 | **Pregunta analítica y datos** | Qué preguntamos, qué datos usamos || 2:00–5:00 | **3 hallazgos clave** | Un gráfico por hallazgo, una frase de interpretación || 5:00–7:00 | **Modelo y métricas** | Qué modelo, qué métrica, qué baseline || 7:00–9:00 | **Demo** | Dashboard o API en vivo (o video si hay riesgo técnico) || 9:00–10:00 | **Limitaciones y siguientes pasos** | Qué harían con más tiempo/datos |**Regla de oro**: *una idea por slide o por bloque*. Si hay dos ideas, son dos slides.

## 4. Principios para los visuales finales- **Fuentes grandes** (mínimo 24pt). Nadie lee un gráfico con letras pequeñas desde atrás.- **Títulos afirmativos**: "El 75% de las mujeres de primera clase sobrevivió" > "Supervivencia por sexo y clase".- **Color con propósito**: gris para el contexto, un color fuerte para lo que quieres destacar.- **Anotaciones inline**: poner el valor encima de la barra importante, no pedirle al público que lea el eje.- **Eliminar chartjunk**: bordes, sombras, cuadrículas innecesarias, leyendas redundantes.

## 5. Utilidad · producir gráficos listos para presentación

```python
import matplotlib.pyplot as pltimport seaborn as snsimport pandas as pd# Configuración global para presentaciónsns.set_theme(style="whitegrid", context="talk")plt.rcParams["font.family"] = "sans-serif"url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"df = pd.read_csv(url)fig, ax = plt.subplots(figsize=(9, 5))agg = df.groupby(["Pclass", "Sex"])["Survived"].mean().reset_index()sns.barplot(x="Pclass", y="Survived", hue="Sex", data=agg, ax=ax)ax.set_title("Las mujeres de 1ra clase sobrevivieron 8× más que los hombres de 3ra",             loc="left", fontsize=14, pad=15)ax.set_xlabel("Clase del pasajero")ax.set_ylabel("Tasa de supervivencia")ax.set_ylim(0, 1)# Anotar los porcentajesfor p in ax.patches:    v = p.get_height()    if v > 0:        ax.annotate(f"{v*100:.0f}%",                    (p.get_x() + p.get_width() / 2., v),                    ha="center", va="bottom", fontsize=11)plt.tight_layout()fig.savefig("hallazgo_principal.png", dpi=200, bbox_inches="tight")plt.show()
```

## 6. Anticipar preguntas del tribunalLas preguntas más frecuentes en defensas de este tipo:- *"¿Cómo validaron el modelo?"* — Tengan la respuesta sobre validación cruzada y métricas a la mano.- *"¿Qué haría con más tiempo / más datos?"* — Mostrar que entienden las limitaciones es señal de madurez.- *"¿Por qué esta métrica y no otra?"* — Siempre justificar la elección con el contexto del caso.- *"¿Podrían usarlo en producción tal como está?"* — Honestidad: qué falta para un despliegue real.- *"¿Cómo se comportan los outliers / valores extremos?"* — Tengan documentadas las decisiones de limpieza.**Respuestas honestas valen más que respuestas pulidas**. Si no saben algo, decir "no lo medimos, pero lo que haríamos sería...".
