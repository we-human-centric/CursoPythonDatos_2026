## Data storytelling

Una presentación analítica **no es un reporte técnico**. No es un volcado de código. Es un **relato** que lleva al público de un problema inicial a una decisión o insight claro.

Los mejores analistas del mundo (Hans Rosling, Edward Tufte, Cole Nussbaumer) coinciden en tres ingredientes:

1. **Contexto**: por qué este problema importa. A quién afecta. Qué está en juego.
2. **Tensión**: qué sorpresa o conflicto reveló el análisis. El momento en que el público se da cuenta de que hay algo importante aquí.
3. **Resolución**: qué acción concreta se propone. Qué debería hacer alguien con esto.

Si después de tu presentación la gente no puede responder esas tres preguntas en 30 segundos, fallaste. No es que no entiendieron. Es que no tuvieron un hilo conductor.

## La pirámide de Minto

Barbara Minto, consultora de McKinsey, desarrolló en los 60 una técnica de comunicación:

```
           Conclusión
          (la punta)
              ↑
       ┌──────┴──────┐
       ↓      ↓      ↓
    Arg1    Arg2   Arg3
       ↓      ↓      ↓
   (evidencia por cada argumento)
```

**Regla**: empieza con la conclusión, apóyala con 3 argumentos, cada uno con evidencia (datos, gráficos).

Lo opuesto es el "suspense académico": cuéntame todo primero, al final te digo qué encontré. **No hagas eso.** Los ejecutivos se aburren. El pulso baja. La gente se va.

## Cómo se ven los gráficos listos para presentación

- **Fuentes grandes**: mínimo 24pt. Nadie lee letra pequeña desde atrás.
- **Títulos afirmativos**: "El 75% de las mujeres de 1ra clase sobrevivió" en lugar de "Supervivencia por sexo y clase".
- **Color con propósito**: gris el contexto, un color fuerte para lo importante.
- **Anotaciones inline**: el valor encima de la barra, no pidiendo que lea el eje.
- **Eliminar ruido**: sin bordes, sin sombras, sin cuadrículas innecesarias, sin leyendas redundantes.

## Producir gráficos listos para defensa

````python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Configuración global para presentación (tamaño de texto más grande)
sns.set_theme(style="whitegrid", context="talk")
plt.rcParams["font.family"] = "sans-serif"

# Cargar datos
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

# Gráfico: supervivencia por clase y sexo
fig, ax = plt.subplots(figsize=(9, 5))
agg = df.groupby(["Pclass", "Sex"])["Survived"].mean().reset_index()
sns.barplot(x="Pclass", y="Survived", hue="Sex", data=agg, ax=ax)

# Título afirmativo (no descriptivo)
ax.set_title(
    "Las mujeres de 1ra clase sobrevivieron 8x más que los hombres de 3ra",
    loc="left", fontsize=14, pad=15
)

# Labels claros
ax.set_xlabel("Clase del pasajero", fontsize=11)
ax.set_ylabel("Tasa de supervivencia", fontsize=11)
ax.set_ylim(0, 1)

# Anotaciones: porcentajes encima de las barras
for p in ax.patches:
    v = p.get_height()
    if v > 0:
        ax.annotate(
            f"{v*100:.0f}%",
            (p.get_x() + p.get_width() / 2., v),
            ha="center", va="bottom", fontsize=11
        )

plt.tight_layout()
fig.savefig("hallazgo_principal.png", dpi=200, bbox_inches="tight")
plt.show()

print("✓ Gráfico listo para presentación. Guardado como hallazgo_principal.png")
````

Ese gráfico es presentable. Texto grande, título claro, números visibles. Un ejecutivo lo mira en 3 segundos y entiende.

## Preguntas del tribunal (y cómo prepararlas)

Las 5 preguntas más frecuentes:

**P: "¿Cómo validaron el modelo?"**
R: Tengan la respuesta sobre validación cruzada, conjuntos de test, qué métrica optimizaron. No digan "pasó los tests". Digan "k-fold con k=5 y Precision/Recall balanceadas".

**P: "¿Qué harían con más tiempo?"**
R: Mostrar que entienden las limitaciones. Mejor que decir "está perfecto". Peor es quedarse en blanco.

**P: "¿Por qué esta métrica y no otra?"**
R: Siempre justificar con el contexto del caso. "Optimizamos por Recall porque es costoso un falso negativo aquí."

**P: "¿Producción real?"**
R: Honestidad. "Así como está, no. Pero con Docker, logs estructurados y monitoreo de drift estaría listo en X semanas".

**P: "¿Outliers?"**
R: Tengan documentadas las decisiones de limpieza. "Removimos el 2% de filas porque [razón]."

**Regla general**: respuestas honestas valen más que respuestas pulidas. Si no saben algo, decir "no lo medimos, pero lo haríamos con..." impresiona más que inventar una respuesta.
